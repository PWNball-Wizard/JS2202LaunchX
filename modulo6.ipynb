#Introducción a las listas.

Acceder a elementos de lista por índice

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

print('The first planet is', planets[0])
print('The second planet is', planets[1])
print('The third planet is', planets[2])
```

```python
planets[3] = 'Red Planet'
print('Mars is also known as', planets[3])
```

#Determinar la longitud de una lista.

```python
number_of_planets = len(planets)
print('There are', number_of_planets, 'planets in the solar system.')
```

#Agregar valores a listas.

```python
planets.append('Pluto')
number_of_planets = len(planets)
print('There are actually', number_of_planets, 'planets in the solar system.')
```

#Eliminar valores de una lista.

```python
planets.pop()  # Goodbye, Pluto
number_of_planets = len(planets)
print('No, there are definitely', number_of_planets, 'planets in the solar system.')
```

#Índices negativos.

```python
print("The first planet is", planets[0])
```

```python
print('The last planet is', planets[-1])
print('The penultimate planet is', planets[-2])
```

#Buscar un valor en una lista.

```python
jupiter_index = planets.index('Jupiter')
print('Jupiter is the', jupiter_index + 1, 'planet from the sun')
```

#Almacenar números en listas.

```python
gravity_on_earth = 1.0
gravity_on_the_moon = 0.166

print(gravity_on_earth)
print(gravity_on_the_moon)
```

```python
gravity_on_planets = [0.378, 0.907, 1, 0.379, 2.36, 0.916, 0.889, 1.12]
bus_weight = 12650 # in kilograms, on Earth

print('On Earth, a double-decker bus weighs', bus_weight, 'kg')
print('On Mercury, a double-decker bus weighs', bus_weight * gravity_on_planets[0], 'kg')
```

#min() y max () con listas.

```python
bus_weight = 12650 # in kilograms, on Earth

print('On Earth, a double-decker bus weighs', bus_weight, 'kg')
print('The lightest a bus would be in the solar system is', bus_weight * min(gravity_on_planets), 'kg')
print('The heaviest a bus would be in the solar system is', bus_weight * max(gravity_on_planets), 'kg')
```

#Manipular datos de lista.

Slice list.

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planets_before_earth = planets[0:2]
print(planets_before_earth)
```

```python
planets_after_earth = planets[3:8]
print(planets_after_earth) 

planets_after_earth = planets[3:]
print(planets_after_earth)
```

#Uniendo listas.

```python
amalthea_group = ['Metis', 'Adrastea', 'Amalthea', 'Thebe']
galilean_moons = ['Io', 'Europa', 'Ganymede', 'Callisto']

regular_satellite_moons = amalthea_group + galilean_moons
print('The regular satellite moons of Jupiter are', regular_satellite_moons)
```

#Ordenar listas.

```python
regular_satellite_moons.sort()
print("The regular satellite moons of Jupiter are", regular_satellite_moons)

regular_satellite_moons.sort(reverse=True)
print("The regular satellite moons of Jupiter are", regular_satellite_moons)
```

#Ejercicio 2: Trabajando con datos de una lista.

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
# Solicitamos el nombre de un planeta *Pista:  input()*
planeta = input("Dame el nombre de un planeta")
# Busca el planeta en la lista
try:
    indice = planets.index(planeta)
    print(f"El planeta {planets[indice]} esta en la lista")
except:
    print("El planeta no esta listado")
# Muestra los planetas más cercanos al sol del que el usuario ingreso
if indice == 0:
    print(f"{planets[indice]} es el planeta mas cercano al sol")
else:
    print(f"Los planetas mas cercanos al sol antes de {planets[indice]} son {planets[0:indice]}")
# Muestra los planetas más lejanos al sol del que el usuario ingreso
print(f"Los planetas mas lejanos al sol despues de {planets[indice]} son {planets[indice + 1:]}")
```
